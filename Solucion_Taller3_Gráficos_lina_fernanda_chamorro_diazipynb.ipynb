la distancia entre Cali y Medellín son 411km. Un vehículo sale de Cali hacia Medellín a 90km/h. Al mismo tiempo sale otro vehiculo de Medellín hacia Cali a 80km/h. Suponiendo una velocidad constante en ambos vehiculos.

¿cuánto tiempo tardan en encontrarse y cuánto han recorrido?

````python
# t = tiempo cuándo se encuentran
# x = distancia desde Cali
e1 = sy.Eq(x+y, 411)
e2 = sy.Eq(x, 90*t)
e3 = sy.Eq(y, 80*t)

solucion = sy.solve(
  (e1,e2,e3),(x,y,t)
)

dist_cali = solucion[x]
dist_mede = solucion[y]
tiempo = solucion[t]

print(f"""
Los dos vehiculos se encuentran pasadas {tiempo:.2f} horas.
Se encuentran a una distancia desde Cali de {dist_cali:.4f} km,
y desde Medellín de {dist_mede:.4f} km.
""")
````

# Discontinuidades

$$ f(x) = \frac{5x}{2x-7} $$

````python
# def f(x):
#   (5*x)/(2*x-7)
#   return
````

````python
f = lambda x: (5*x)/(2*x - 7)
````

````python
ss = sy.singularities(f(x), x)
print("La ecuación f(x) =")
display(f(x))
print("\ndiscontinuidad en:")
display(ss)
````

````python
xs = np.linspace(-2, 8, 100)
ys = f(xs)
# encontrar las discontinuidades = singularities
````

````python
fig, ax = plt.subplots()
ax.plot(xs, ys)
for xd in list(ss):
  ax.axvline(x=xd.evalf(), color='green')
plt.grid()
plt.show()
````

# Raices de una función

$$ f(x) = x^2 - 3x - 6 $$

````python
e1 = sy.Eq(x**2 - 3*x - 6, 0)
raices = sy.solve((e1),(x,y))
raices
````

````python
def f(x):
  return x**2 - 3*x - 6
````

````python
e1 = sy.Eq(f(x), 0)
raices = sy.solve(e1, x)
raices
````

````python
fig, ax = plt.subplots()
xs = np.linspace(-10,10,100)
ys = f(xs)
ax.axhline(0, color='k') # colores: rgbcmywk
# ciclo para pintar las raices
for x_raiz in raices:
  ax.plot(x_raiz, f(x_raiz), 'ro')
ax.plot(xs, ys)
````
